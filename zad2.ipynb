```
import numpy as np
import pandas as pd
import sklearn
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas_datareader.data as web
from sklearn.model_selection import train_test_split


sns.set()
 
 
# generowanie danych
data = {
    'size': [np.nan, 'L', 'M', 'L', 'S', np.nan, 'L', 'S', 'L', 'M', 'S', 'XL', 'M', 'S', 'L', 'S', 'M', 'L', 'S', 'L'],
    'color': [np.nan, 'green', 'blue', 'green', 'red', 'yellow', np.nan, 'blue', 'green', 'black', 'blue', 'yellow', 'green', 'green', 'black', 'red', 'yellow', 'grey', 'blue', 'gray'],
    'gender': ['male', 'male', 'male', np.nan, np.nan, 'female', np.nan, 'male', 'male', 'male', 'male', np.nan, 'female', 'female', 'female', 'male', 'female', 'male', 'male', 'male'],
    'price': [np.nan, 89.0, 99.0, 129.0, 79.0, 230.0, np.nan, 180.0, np.nan, 80.0, np.nan, 189.0, 79.0, 179.0, 29.0, 130.0, 180.0, 110.0, 46.0, 89.0],
    'weight': [ 450, 300, np.nan, 410, np.nan, 700, 630, np.nan, 300, 240, 320, 450, 400, 630, 240, 100, 780, 610, 660, 300],
    'bought': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'yes', 'no', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'yes', 'no', 'no']
}
 
df_raw = pd.DataFrame(data=data)
df_raw
 
df = df_raw.copy()
#zad1
df.dropna(thresh=len(df.keys())-2, inplace=True)
print(df.isnull().sum())
#zad2
si = SimpleImputer(missing_values=np.nan, strategy='mean')
df['price'] = si.fit_transform(df[['price']])
print(df)

#zad3
df['weight'] = pd.Series(df['weight']).interpolate() 
print(df)

#zad4 
si = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='unisex')
df['size'] = si.fit_transform(df[['size']])
si = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='black')
df['color'] = si.fit_transform(df[['color']])
si = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='male')
df['gender'] = si.fit_transform(df[['gender']])
print(df)

#zad5
df['cost'] = pd.cut(df['price'], bins = 3, labels=('cheap', 'normal_price', 'expensive'))
print(df)
#zad6
df['bought'] = df['bought'].apply(lambda x: 1 if 'yes' in x else 0)
print(df)
df_in = df[['size', 'color', 'gender', 'price', 'weight', 'cost']]
df_out = df['bought']

X_train, X_test, Y_train, Y_test = train_test_split(df_in, df_out)
print("X_train","\n", X_train)
print("X_test", "\n", X_test)
print("Y_train", "\n", Y_train)
print("Y_test", "\n", Y_test)
print(Y_train.head())
```
