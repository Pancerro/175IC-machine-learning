```

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
data = load_iris()
input = data['data'][:, :2]
output = data['target'].astype('int')
x_train, x_test, y_train, y_test = train_test_split(input, output)
svc = SVC(C=0.3721, kernel='linear')
svc.fit(x_train, y_train)
plot_decision_regions(x_train, y_train, clf=svc)
accuracy = svc.score(x_test, y_test)
plt.title(f"SVC: train accuracy: {accuracy}")
plt.xlabel('petal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
svc = SVC(C=0.2137, kernel='poly')
svc.fit(x_train, y_train)
plot_decision_regions(x_train, y_train, clf=svc)
accuracy = svc.score(x_test, y_test)
plt.title(f"SVC: train accuracy: {accuracy}")
plt.xlabel('petal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
svc = SVC(C=0.7312, kernel='rbf')
svc.fit(x_train, y_train)
plot_decision_regions(x_train, y_train, clf=svc)
accuracy = svc.score(x_test, y_test)
plt.title(f"SVC: train accuracy: {accuracy}")
plt.xlabel('petal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
svc = SVC(C=0.1327, kernel='sigmoid')
svc.fit(x_train, y_train)
plot_decision_regions(x_train, y_train, clf=svc)
accuracy = svc.score(x_test, y_test)
plt.title(f"SVC: train accuracy: {accuracy}")
plt.xlabel('petal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()

```
