```
import pandas as pd
import plotly.express as px
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans

data, _ = make_blobs(n_samples=1000, centers=4, random_state=11)
df = pd.DataFrame(data, columns=['x1', 'x2'])
px.scatter(df, 'x1', 'x2')
cluster = DBSCAN(eps=0.6, min_samples=6)
cluster.fit(data)
df['cluster'] = cluster.labels_
fig = px.scatter(df, 'x1', 'x2', 'cluster', color_continuous_midpoint=0)
fig.show()

X = df[['x1', 'x2']]
k_means = KMeans(n_clusters=4, random_state=11).fit(X)
df['k_means'] = k_means.labels_
fig2 = px.scatter(df, 'x1', 'x2', 'k_means')
fig2.show()

def equals_cluster():
  dbscan_counts = df['cluster'].value_counts()
  k_means_counts = df['k_means'].value_counts()
  print("DBSCAN:")
  print(dbscan_counts)
  print("KMEANS:")
  print(k_means_counts)

equals_cluster()
```
