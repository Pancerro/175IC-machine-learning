```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.ensemble import IsolationForest

data = make_blobs(n_samples=600, cluster_std=2.0, random_state=10)[0]

forest = IsolationForest(n_estimators=300, contamination=0.05)
predict = forest.fit_predict(data)
all_data = pd.DataFrame(data, columns=['x', 'y'])
all_data['outliner'] = predict
 
plt.figure(figsize=(10, 5))
inliners = all_data[all_data.outliner == 1].drop(columns='outliner')
outliners = all_data[all_data.outliner == -1].drop(columns='outliner')
  
plt.scatter(x=inliners.x, y=inliners.y, edgecolors='blue')
plt.scatter(x=outliners.x, y=outliners.y, edgecolors='red')
plt.show()
```
